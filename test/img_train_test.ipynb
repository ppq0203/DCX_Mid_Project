```python
import tensorflow as tf
import numpy as np

from keras import Sequential
from keras.layers import Dense, Input

model = Sequential()
# Flatten : 신경망에 데이터를 넣어줄때 1차원으로 쭉 펴서 넣어주는 함수
model.add(Input(shape=150528))

model.add(Dense(1000, activation='sigmoid'))
model.add(Dense(750, activation='sigmoid'))
model.add(Dense(500, activation='sigmoid'))
model.add(Dense(200, activation='sigmoid'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(50, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
# 출력층 (정답이 0~9까지 10개이므로 뉴런은 10개, 다중분류에는 softmax가 오차의 총합을 줄이는데 더 효율적)
model.add(Dense(2, activation='softmax'))
model.summary()
```

```python
%cd /content/drive/MyDrive/dcx_mid_project/DCX_Mid_Project/test

# food = np.load('food64.npy')
food = np.load('./result.npy')
np.random.shuffle(food)
x = food[:,:-2]
y = food[:,-2:]
print(food.shape, x.shape, y.shape)
train_num = 1500
x_train = x[:train_num, :]
x_test = x[train_num:, :]
y_train = y[:train_num, :]
y_test = y[train_num:, :]
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["acc"]
              )

```

```python
h4 = model.fit(x_train, y_train,
              epochs=5,
              batch_size = 128
              )
```

```python
model.evaluate(x_test, y_test)
```
